```python
# Calculating yearly growth rates for GVA
def gva_growth(data, column_prefix, start_year, end_year):
    for year in range(start_year, end_year):
        current_year_col = f'{column_prefix} {year}'
        next_year_col = f'{column_prefix} {year + 1}'
        growth_rate_col = f'GVA Growth {year + 1}'
        
        data[growth_rate_col] = (data[next_year_col] - data[current_year_col]) / data[current_year_col]
    
    return data

df = gva_growth(df, 'Total Company GVA', 2017, 2021)
```

```python
# Train, Test and Validation splits
train_list = []
val_list = []
test_list = []

# Getting company names to split individually for each company
company_names = df_panel['Company Name'].unique()

# Looping over each company to split data into training, test, and validation
for company_name in company_names:
    company_data = df_panel[df_panel['Company Name'] == company_name]
    num_rows = company_data.shape[0]
    print(f"Company Name: {company_name}, Total Rows: {num_rows}")
    
    if num_rows > 2:  # If the number of time series > 2, data is split into train (70%), test (15%), and validation (15%)
        company_data = company_data.sort_values(by='Year')
        train_split_index = int(0.7 * num_rows)
        val_split_index = int(0.85 * num_rows)
        
        train_data = company_data.iloc[:train_split_index]
        val_data = company_data.iloc[train_split_index:val_split_index]
        test_data = company_data.iloc[val_split_index:]
        
        train_list.append(train_data)
        val_list.append(val_data)
        test_list.append(test_data)
    elif num_rows > 1:  # Only splitting into training and testing if there are only 2 time series rows
        company_data = company_data.sort_values(by='Year')
        split_index = int(0.8 * num_rows)
        train_data = company_data.iloc[:split_index]
        test_data = company_data.iloc[split_index:]
        train_list.append(train_data)
        test_list.append(test_data)
    else:
        # Adding companies with only one row directly in the training set
        train_list.append(company_data)

train_df = pd.concat(train_list)
val_df = pd.concat(val_list)
test_df = pd.concat(test_list)
```

```python
#defining the minmaxscaler for normalizing data and the function for preparing lstm data using defined time steps

scaler = MinMaxScaler(feature_range=(0, 1))
column_transformer = ColumnTransformer(
    transformers=[
        ('scaler', scaler, ['Employment', 'GVA'])
    ],
    remainder='passthrough'  # Keep the other columns unchanged
)


def prepare_lstm_data(data, time_steps=1):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps)])
        y.append(data[i + time_steps])
    return np.array(X), np.array(y)
```

```python
# Defining GVA as the target variable
train_values = train_df['GVA'].values
print(f"Original train_values shape: {train_values.shape}")

# Applying the scaler
train_values_scaled = scaler.fit_transform(train_values.reshape(-1, 1))
print(f"Scaled train_values shape: {train_values_scaled.shape}")
```

```python
# Preparing data using pre-defined function for LSTM with a single time step
time_steps = 1
X_train, y_train = prepare_lstm_data(train_values_scaled, time_steps)
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
```

```python
# Preparing validation data similarly
val_values = val_df['GVA'].values
val_values_scaled = scaler.transform(val_values.reshape(-1, 1))
X_val, y_val = prepare_lstm_data(val_values_scaled, time_steps)
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
```

```python
# Plotting the original vs scaled GVA data to check if scaling worked properly

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.hist(train_values, bins=30, alpha=0.7, label='original GVA')
plt.legend()
plt.title('The original GVA data')
plt.yscale('log')  # Using log scale for better visibility

plt.subplot(2, 1, 2)
plt.hist(train_values_scaled, bins=30, alpha=0.7, label='scaled GVA')
plt.legend()
plt.title('The scaled GVA data')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
